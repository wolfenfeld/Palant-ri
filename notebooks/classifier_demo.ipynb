# Classifier Plots

## Relevant Imports

```python
import sys

import pandas as pd

from sklearn import svm
from sklearn.datasets import load_iris, load_breast_cancer

if '../' not in sys.path:
    sys.path.append('../')
```

# Binary Labled Data - Breast Cancer Dataset

```python
breast_cancer = load_breast_cancer()

breast_cancer_df = pd.DataFrame(breast_cancer.data[:, :3],columns=['x1','x2','x3'])
breast_cancer_df['y'] = breast_cancer.target

breast_cancer_clf = svm.SVC(kernel='rbf',probability=True)
breast_cancer_clf.fit(breast_cancer_df[['x1','x2','x3']], breast_cancer_df['y']);
```

# Multiclass Labled Data - Iris Dataset

```python
iris = load_iris()

iris_df = pd.DataFrame(iris.data[:, :3],columns=['x1','x2','x3'])
iris_df['y'] = iris.target

iris_clf = svm.SVC(kernel='rbf',probability=True)
iris_clf.fit(iris_df[['x1','x2']], iris_df['y']);
```

# Two Dimensional Classifier

```python
from palantiri.ClassifierPlotHandlers import TwoDimensionalClassifierPlotHandler

plot_handler = TwoDimensionalClassifierPlotHandler(iris_df[['x1','x2','y']], iris_clf)
```

```python
plot_handler.plot_prediction()
```

```python
plot_handler.plot_roc()
```

```python
plot_handler.plot_confusion_matrix()
```

# Three Dimensional Classifier

```python
from palantiri.ClassifierPlotHandlers import ThreeDimensionalClassifierPlotHandler

plot_handler = ThreeDimensionalClassifierPlotHandler(breast_cancer_df[['x1','x2','x3','y']], breast_cancer_clf)
```

```python
plot_handler.plot_prediction()
```
